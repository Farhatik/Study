## Задание 1. Реализация градиентного спуска (3.5 балла)

В этом задании вам предстоит написать собственные реализации различных подходов к градиентному спуску с опорой на подготовленные шаблоны в файле  `utils.py`:

**Задание 1.1. (0.5 балла)** Полный градиентный спуск **GradientDescent**:

$$
    w_{k + 1} = w_{k} - \eta_{k} \nabla_{w} Q(w_{k}).
$$

**Задание 1.2. (1 балл)** Стохастический градиентный спуск **StochasticDescent**:

$$
    w_{k + 1} = w_{k} - \eta_{k} \nabla_{w} q_{i_{k}}(w_{k}).
$$ 

$\nabla_{w} q_{i_{k}}(w_{k}) \,$ - это оценка градиента по одному объекту, выбранному случайно.

**Задание 1.3. (1 балл)** Метод Momentum **MomentumDescent**:

$$
    h_0 = 0, \\
    h_{k + 1} = \alpha h_{k} + \eta_k \nabla_{w} Q(w_{k}), \\
    w_{k + 1} = w_{k} - h_{k + 1}.
$$

**Задание 1.4. (1 балл)** Метод Adaptive gradient algorithm **Adagrad**:

$$
    G_0 = 0, \\
    G_{k + 1} = G_{k} + \left(\nabla_{w} Q(w_{k})\right) ^ 2, \\
    w_{k + 1} = w_{k} - \dfrac{\eta_k}{\sqrt{\varepsilon + G_{k}}} \nabla_{w} Q(w_{k}).
$$


Во всех вышеназванных методах мы будем использовать следующую формулу для длины шага:

$$
    \eta_{k} = \lambda \left(\dfrac{s_0}{s_0 + k}\right)^p
$$
На практике достаточно настроить параметр $\lambda$, а остальным выставить параметры по умолчанию: $s_0 = 1, \, p = 0.5.$

Мы будем использовать функцию потерь MSE:

$$
    Q(w) = \dfrac{1}{\ell} \sum\limits_{i=1}^{\ell} (a_w(x_i) - y_i)^2
$$

Все вычисления должны быть векторизованы.

```python
num_objects = 100
dimension = 5

X = np.random.rand(num_objects, dimension)
y = np.random.rand(num_objects)

lambda_ = 1e-2
w0 = np.zeros(dimension)

max_iter = 10
tolerance = 0
```

```python
# GradientDescent

descent = GradientDescent(lambda_ = lambda_, w0 = w0)

gradient = descent.calc_gradient(X, y)

assert gradient.shape[0] == dimension, 'Gradient failed'

diff = descent.step(X, y, 0)

assert diff.shape[0] == dimension, 'Weights failed'
```

```python
# StochasticDescent

descent = StochasticDescent(lambda_ = lambda_, w0 = w0)

gradient = descent.calc_gradient(X, y)

assert gradient.shape[0] == dimension, 'Gradient failed'

diff = descent.step(X, y, 0)

assert diff.shape[0] == dimension, 'Weights failed'
```

```python
# MomentumDescent

descent = MomentumDescent(lambda_ = lambda_, w0 = w0)

gradient = descent.calc_gradient(X, y)

assert gradient.shape[0] == dimension, 'Gradient failed'

diff = descent.step(X, y, 0)

assert diff.shape[0] == dimension, 'Weights failed'
```

```python
# Adagrad

descent = Adagrad(lambda_ = lambda_, w0 = w0)

gradient = descent.calc_gradient(X, y)

assert gradient.shape[0] == dimension, 'Gradient failed'

diff = descent.step(X, y, 0)

assert diff.shape[0] == dimension, 'Weights failed'
```

```python
# LinearRegression

regression = LinearRegression(
    descent = StochasticDescent(lambda_ = lambda_, w0 = w0, batch_size = 2),
    tolerance = tolerance,
    max_iter = max_iter
)

regression.fit(X, y)

assert len(regression.loss_history) == max_iter, 'Loss history failed'

prediction = regression.predict(X)

assert prediction.shape[0] == num_objects, 'Predict failed'
```

## Задание 5. Сравнение методов градиентного спуска (2 балла)

В этом задании вам предстоит сравнить методы градиентного спуска на подготовленных вами данных из предыдущего задания.

* **Задание 5.1. (1.5 балла)** Подберите по валидационной выборке наилучшую длину $\lambda$ шага для каждого метода. Для этого можно сделать перебор по логарифмической сетке, так как нас интересует скорее порядок величины, нежели её точное значение. Сравните качество методов по метрикам MSE и R^2 на обучающей и тестовой выборках, сравните количество итераций до сходимости. Все параметры кроме $\lambda$ стоит выставить равным значениям по умолчанию.

* **Задание 5.2. (0.5 балла)** Постройте график зависимости значения функции ошибки от номера итерации (все методы на одном графике).

Посмотрите на получившиеся результаты. Сравните методы между собой.

```python
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
X_train = train.drop(columns=['log_price'])
y_train = train.log_price

X_test = test.drop(columns=['log_price'])
y_test = test['log_price']

X_valid = valid.drop(columns=['log_price'])
y_valid = valid.log_price

column_transformer = ColumnTransformer([
    ('ohe', OneHotEncoder(handle_unknown="ignore"), categorical),
    ('scaling', StandardScaler(), numeric_features)
],  sparse_threshold=0)

w0 = np.zeros(column_transformer.fit_transform(X_train).shape[1])
```

```python
lambdas = np.logspace(-3, -1, 6)

dfl_ = pd.DataFrame(data=np.tile([1e-2, 10000, 1001, 10000, [], 2], (4, 1)), 
                    index=["Adagrad", "GradientDescent", "MomentumDescent", "StochasticDescent"],
                    columns=['lambda_', "val_mse", "num_iter", "test_mse", "r2", "loss_history"]
                   )

for lamda_ in lambdas:
#     print("lambda =", lamda_)
    regression0 = LinearRegression(descent = Adagrad(lambda_ = lamda_, w0 = w0))
    regression1 = LinearRegression(descent = GradientDescent(lambda_ = lamda_, w0 = w0))
    regression2 = LinearRegression(descent = MomentumDescent(lambda_ = lamda_, w0 = w0))
    regression3 = LinearRegression(descent = StochasticDescent(lambda_ = lamda_, w0 = w0))
    
    X_scaled_train = column_transformer.fit_transform(X_train)
    X_scaled_test = column_transformer.transform(X_test)
    X_scaled_valid =  column_transformer.transform(X_valid)
    
    model0 = regression0.fit(X_scaled_train, y_train)
    y_pred0v = model0.predict(X_scaled_valid)
    y_pred0t = model0.predict(X_scaled_test)
    
    mse0v = mean_squared_error(y_pred0v, y_valid)
    if mse0v < dfl_.loc["Adagrad", "val_mse"]:
        dfl_.loc["Adagrad", "val_mse"] = mse0v
        dfl_.loc["Adagrad", "lambda_"] = lamda_
        dfl_.loc["Adagrad", "num_iter"] = len(regression0.loss_history)
        dfl_.loc["Adagrad", "loss_history"] = regression0.loss_history
        dfl_.loc["Adagrad", "test_mse"] = mean_squared_error(y_pred0t, y_test)
        dfl_.loc["Adagrad", "r2"] = r2_score(y_pred0t, y_test)
        
#     print("Adograd, mse_valid =", mse0v)
    model1 = regression1.fit(X_scaled_train, y_train)
    y_pred1v = model1.predict(X_scaled_valid)
    y_pred1t = model1.predict(X_scaled_test)
    mse1v = mean_squared_error(y_pred1v, y_valid)
    if mse1v < dfl_.loc["GradientDescent", "val_mse"]:
        dfl_.loc["GradientDescent", "val_mse"] = mse1v
        dfl_.loc["GradientDescent", "lambda_"] = lamda_
        dfl_.loc["GradientDescent", "num_iter"] = len(regression1.loss_history)
        dfl_.loc["GradientDescent", "loss_history"] = regression1.loss_history
        dfl_.loc["GradientDescent", "test_mse"] = mean_squared_error(y_pred1t, y_test)
        dfl_.loc["GradientDescent", "r2"] = r2_score(y_pred1t, y_test)
#     print("GradientDescent, mse =", mse1)

    model2 = regression2.fit(X_scaled_train, y_train)
    y_pred2v = model2.predict(X_scaled_valid)
    y_pred2t = model2.predict(X_scaled_test)
    mse2v = mean_squared_error(y_pred2v, y_valid)
    if mse2v < dfl_.loc["MomentumDescent", "val_mse"]:
        dfl_.loc["MomentumDescent", "val_mse"] = mse2v
        dfl_.loc["MomentumDescent", "lambda_"] = lamda_
        dfl_.loc["MomentumDescent", "num_iter"] = len(regression2.loss_history)
        dfl_.loc["MomentumDescent", "loss_history"] = regression2.loss_history
        dfl_.loc["MomentumDescent", "test_mse"] = mean_squared_error(y_pred2t, y_test)
        dfl_.loc["MomentumDescent", "r2"] = r2_score(y_pred2t, y_test)
        
#     print("MomentumDescent, mse =", mse2)
    model3 = regression3.fit(X_scaled_train, y_train)
    y_pred3v = model3.predict(X_scaled_valid)
    y_pred3t = model3.predict(X_scaled_test)
    mse3v = mean_squared_error(y_pred3v, y_valid)
    if mse3v < dfl_.loc["StochasticDescent", "val_mse"]:
        dfl_.loc["StochasticDescent", "val_mse"] = mse3v
        dfl_.loc["StochasticDescent", "lambda_"] = lamda_
        dfl_.loc["StochasticDescent", "num_iter"] = len(regression3.loss_history)
        dfl_.loc["StochasticDescent", "loss_history"] = regression3.loss_history
        dfl_.loc["StochasticDescent", "test_mse"] = mean_squared_error(y_pred3t, y_test)
        dfl_.loc["StochasticDescent", "r2"] = r2_score(y_pred3t, y_test)
#     print("StochasticDescent, mse=", mse3)
#     print(dfl_)
#     print("***************")
```

Видно, что лучше всего MSE у MomentumDescent, а наилучшее r2 - MomentumDescent, быстрее всего сошелся MomentumDescent

В общем все видно в табличке

```python
plt.figure(figsize=(14, 10))
plt.title("график зависимости значения функции ошибки от номера итерации")
plt.ylabel("Значение MSE")
plt.xlabel("Номер итерации")
ax1 = plt.plot(np.arange(dfl_.loc["Adagrad", "num_iter"]),
         dfl_.loc["Adagrad", "loss_history"])
ax2 = plt.plot(np.arange(dfl_.loc["GradientDescent", "num_iter"]),
         dfl_.loc["GradientDescent", "loss_history"])
ax3 = plt.plot(np.arange(dfl_.loc["MomentumDescent", "num_iter"]),
         dfl_.loc["MomentumDescent", "loss_history"])

ax4 = plt.plot(np.arange(dfl_.loc["StochasticDescent", "num_iter"]),
         dfl_.loc["StochasticDescent", "loss_history"])
plt.legend(["Adagrad", "GradientDescent", "MomentumDescent", "StochasticDescent"],  prop={'size': 20})

plt.show()
```

```python
import time
batch_sizes = np.arange(5, 500, 10)

X_scaled_train = column_transformer.fit_transform(X_train)
X_scaled_test = column_transformer.fit_transform(X_test)
info = list()
for k in range(5):
    print("number of attempt =", k)
    qty_of_steps = list()
    times = list()
    for batch_size in batch_sizes:
        t1 = time.time()
        regression = LinearRegression(
            descent = StochasticDescent(lambda_ = 1,
                                        w0 = w0,
                                        batch_size=batch_size),
            max_iter = 1000
        )
        regression.fit(X_scaled_train, y_train)
        t2 = time.time()
        qty_of_steps.append(len(regression.loss_history))
        times.append(t2 - t1)
        print("batch_size = ", batch_size, "\t time =", t2 - t1, "\t qty_of_iters =", len(regression.loss_history))
    print("*******************************************************")
    info.append([qty_of_steps, times])
```

```python
infoar = np.array(info)
```

## Задание 7. Регуляризация (1 балл)

В этом задании вам предстоит исследовать влияние регуляризации на работу различных методов градиентного спуска. (Напомним, регуляризация - это добавка к функции потерь, которая штрафует за норму весов). Мы будем использовать l2 регуляризацию, таким образом функция потерь приобретает следующий вид:

$$
    Q(w) = \dfrac{1}{\ell} \sum\limits_{i=1}^{\ell} (a_w(x_i) - y_i)^2 + \dfrac{1}{2} \| w \|^2
$$

Допишите классы **GradientDescentReg**, **StochasticDescentReg**, **MomentumDescentReg**, **AdagradReg** в файле `utils.py`. Мы будем использовать тот же самый класс для линейной регрессии, так как для сравнения методов с регуляризацией и без неё нам нужна только MSE часть функции потерь.

Найдите по валидационной выборке лучшие параметры обучения с регуляризацией. Сравните для каждого метода результаты на тестовой выборке по метрикам MSE и R^2 с регуляризацией и без регуляризации. Постройте для каждого метода график со значениями функции потерь MSE с регуляризацией и без регуляризации.

Посмотрите на получившиеся результаты. Какие можно сделать выводы, как регуляризация влияет на сходимость? Чем вы можете объяснить это?

```python
%autoreload 2

from utils import (
    AdagradReg,
    GradientDescentReg,
    MomentumDescentReg,
    StochasticDescentReg,
)

mues = np.logspace(-3, -1, 6)

dfm_ = pd.DataFrame(data=np.tile([1e-2, 10000, 1001, 10000, 2, []], (8, 1)), 
                    index=["AdagradReg", "GradientDescentReg", "MomentumDescentReg", 
                           "StochasticDescentReg", "Adagrad", "GradientDescent", "MomentumDescent", "StochasticDescent"],
                    columns=['mu', "val_mse", "num_iter", "test_mse", "r2", "loss_history"]
                   )

for mu in mues:
#     print("lambda =", lamda_)
    regression0 = LinearRegression(descent = AdagradReg(lambda_ = 0.5, w0 = w0, mu=mu))
    regression1 = LinearRegression(descent = GradientDescentReg(lambda_ = 0.5, w0 = w0, mu=mu))
    regression2 = LinearRegression(descent = MomentumDescentReg(lambda_ = 0.5, w0 = w0, mu=mu))
    regression3 = LinearRegression(descent = StochasticDescentReg(lambda_ = 0.5, w0 = w0, mu=mu))
    
    X_scaled_train = column_transformer.fit_transform(X_train)
    X_scaled_test = column_transformer.transform(X_test)
    X_scaled_valid =  column_transformer.transform(X_valid)
    
    model0 = regression0.fit(X_scaled_train, y_train)
    y_pred0v = model0.predict(X_scaled_valid)
    y_pred0t = model0.predict(X_scaled_test)
    
    mse0v = mean_squared_error(y_pred0v, y_valid)
    if mse0v < dfm_.loc["AdagradReg", "val_mse"]:
        dfm_.loc["AdagradReg", "val_mse"] = mse0v
        dfm_.loc["AdagradReg", "mu"] = mu
        dfm_.loc["AdagradReg", "num_iter"] = len(regression0.loss_history)
        dfm_.loc["AdagradReg", "loss_history"] = regression0.loss_history
        dfm_.loc["AdagradReg", "test_mse"] = mean_squared_error(y_pred0t, y_test)
        dfm_.loc["AdagradReg", "r2"] = r2_score(y_pred0t, y_test)
        
    model1 = regression1.fit(X_scaled_train, y_train)
    y_pred1v = model1.predict(X_scaled_valid)
    y_pred1t = model1.predict(X_scaled_test)
    mse1v = mean_squared_error(y_pred1v, y_valid)
    if mse1v < dfm_.loc["GradientDescentReg", "val_mse"]:
        dfm_.loc["GradientDescentReg", "val_mse"] = mse1v
        dfm_.loc["GradientDescentReg", "mu"] = mu
        dfm_.loc["GradientDescentReg", "num_iter"] = len(regression1.loss_history)
        dfm_.loc["GradientDescentReg", "loss_history"] = regression1.loss_history
        dfm_.loc["GradientDescentReg", "test_mse"] = mean_squared_error(y_pred1t, y_test)
        dfm_.loc["GradientDescentReg", "r2"] = r2_score(y_pred1t, y_test)

    model2 = regression2.fit(X_scaled_train, y_train)
    y_pred2v = model2.predict(X_scaled_valid)
    y_pred2t = model2.predict(X_scaled_test)
    mse2v = mean_squared_error(y_pred2v, y_valid)
    if mse2v < dfm_.loc["MomentumDescentReg", "val_mse"]:
        dfm_.loc["MomentumDescentReg", "val_mse"] = mse2v
        dfm_.loc["MomentumDescentReg", "mu"] = mu
        dfm_.loc["MomentumDescentReg", "num_iter"] = len(regression2.loss_history)
        dfm_.loc["MomentumDescentReg", "loss_history"] = regression2.loss_history
        dfm_.loc["MomentumDescentReg", "test_mse"] = mean_squared_error(y_pred2t, y_test)
        dfm_.loc["MomentumDescentReg", "r2"] = r2_score(y_pred2t, y_test)
        
    model3 = regression3.fit(X_scaled_train, y_train)
    y_pred3v = model3.predict(X_scaled_valid)
    y_pred3t = model3.predict(X_scaled_test)
    mse3v = mean_squared_error(y_pred3v, y_valid)
    if mse3v < dfm_.loc["StochasticDescentReg", "val_mse"]:
        dfm_.loc["StochasticDescentReg", "val_mse"] = mse3v
        dfm_.loc["StochasticDescentReg", "mu"] = mu
        dfm_.loc["StochasticDescentReg", "num_iter"] = len(regression3.loss_history)
        dfm_.loc["StochasticDescentReg", "loss_history"] = regression3.loss_history
        dfm_.loc["StochasticDescentReg", "test_mse"] = mean_squared_error(y_pred3t, y_test)
        dfm_.loc["StochasticDescentReg", "r2"] = r2_score(y_pred3t, y_test)
```
